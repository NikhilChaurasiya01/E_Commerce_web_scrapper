```python
import pandas as pd

# Step 1: Load the CSV file
df = pd.read_csv('products.csv')

# Step 2: Replace all NaN values with 0
df_filled = df.fillna(0)

# Step 3: Save the updated DataFrame to a new CSV file
df_filled.to_csv('products_filled.csv', index=False)

print("NaN values replaced with 0 and saved as 'products_filled.csv'")
```
